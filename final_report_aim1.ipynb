```python
# run_mrg_etiv_phaseorder_compare_pA_cv_polarity_matched.py
"""
Compare CV (ETI-filtered voltage->current) vs biphasic CC (charge-balanced)
thresholds on MRG-like axon models with diameter-dependent gnabar scaling.

This version:
 - for each CC phase order ('anodic'/'cathodic') uses a matching CV polarity
 - Ith values reported in pA
 - robust threshold search with diagnostic ap_stim/ap_det
"""
import os
import sys
import traceback
import numpy as np
import matplotlib.pyplot as plt
from neuron import h, gui

LIB_PATH = "arm64/.libs/libnrnmech.so"

V0 = 1e-2
t_on = 0.0002
t_off = 0.0007
I_target_A = 100e-6
DT_SIM = 0.02
detect_node_idx = 10

DIAMETERS = [0.8, 1.5, 2.0, 5.7, 6.0, 7.0, 8.0, 9.0, 10.0, 11.0, 12.0 , 13.0, 14.0, 15.0, 16.0
ALPHAS = [0.8]

t_total = 0.002
dt_wave = 1e-6

BIP_DUR_CATH = 500e-6
BIP_DUR_ANOD = 500e-6
BIP_IPG = 0.0

PHASE_ORDERS = ['anodic', 'cathodic']

INTERNODE_CAP_UM = 500.0
FIXED_INTERNODE_FOR_DEBUG = None
THRESH_MAX_FACTOR = 1e4
THRESH_TOL = 1e-3
THRESH_BINARY_ITERS = 40

h.dt = DT_SIM

def load_mech_lib_safe(lib_path, probe_mech_names=('axnode','AXNODE','mrgNa','node')):
    if lib_path is None:
        print("[load_mech_lib_safe] LIB_PATH is None — skipping mech load.")
        return True
    lib_path = os.path.abspath(lib_path)
    print(f"[load_mech_lib_safe] lib_path -> {lib_path}")
    if not os.path.exists(lib_path):
        raise FileNotFoundError(f"Mechanism library not found at {lib_path}")
    try:
        for mech in probe_mech_names:
            try:
                sec = h.Section(name='_probe_')
                try:
                    sec.insert(mech)
                    try:
                        del sec
                    except Exception:
                        pass
                    print(f"[probe] mechanism '{mech}' INSERT succeeded -> skip nrn_load_dll.")
                    return True
                except Exception:
                    try:
                        del sec
                    except Exception:
                        pass
            except Exception:
                pass
    except Exception:
        pass
    try:
        h.nrn_load_dll(lib_path)
        print(f"[load_mech_lib_safe] SUCCESS: loaded mechanisms from {lib_path}")
        return True
    except Exception as e:
        msg = str(e)
        already_patterns = ['user defined name already exists', 'already exists', 'hocobj_call error']
        if any(pat in msg for pat in already_patterns):
            print("[load_mech_lib_safe] Warning: library load reported 'already exists'. Treating as OK.")
            print("  err (truncated):", msg.splitlines()[:4])
            return True
        traceback.print_exc()
        raise

def rectangular_pulse(V0_local, t_on_local, t_off_local):
    def Vfunc(t):
        return V0_local if (t >= t_on_local and t <= t_off_local) else 0.0
    return Vfunc

def biphasic_current_waveform(I_cath_A,
                              dur_cath_s,
                              dur_anod_s=None,
                              ipg_s=0.0,
                              tvec=None,
                              start_time=None,
                              phase_first='anodic'):
    if tvec is None:
        raise ValueError("tvec must be provided")
    if dur_anod_s is None:
        dur_anod_s = dur_cath_s
    if start_time is None:
        start_time = t_on
    I_anod_amp = abs(I_cath_A) * (dur_cath_s / dur_anod_s)
    I = np.zeros_like(tvec)
    if phase_first.lower().startswith('c'):
        t1_start = start_time
        t1_end = t1_start + dur_cath_s
        t2_start = t1_end + ipg_s
        t2_end = t2_start + dur_anod_s
        I[(tvec >= t1_start) & (tvec < t1_end)] = -abs(I_cath_A)
        I[(tvec >= t2_start) & (tvec < t2_end)] = +abs(I_anod_amp)
    else:
        t1_start = start_time
        t1_end = t1_start + dur_anod_s
        t2_start = t1_end + ipg_s
        t2_end = t2_start + dur_cath_s
        I[(tvec >= t1_start) & (tvec < t1_end)] = +abs(I_anod_amp)
        I[(tvec >= t2_start) & (tvec < t2_end)] = -abs(I_cath_A)
    return I, (t1_start, t1_end, t2_start, t2_end)

def generate_eti_current(V_app_func, tvec, R_s=100.0, R_ct=1e6, C_dl=1e-6):
    dt = float(tvec[1] - tvec[0])
    N = len(tvec)
    Vp = np.zeros(N); I = np.zeros(N)
    for n in range(N-1):
        V_app_next = V_app_func(tvec[n+1])
        a = C_dl/dt + 1.0/R_ct + 1.0/R_s
        b = C_dl*Vp[n]/dt + V_app_next/R_s
        Vp_next = b / a
        I_next = (V_app_next - Vp_next)/R_s
        Vp[n+1] = Vp_next; I[n+1] = I_next
    return I

def prepare_waveform_for_neuron(t_s, I_A, I_target_A=I_target_A, hard_clip_A=200e-6):
    t_ms = t_s * 1e3
    I_nA = I_A * 1e9
    peak_nA = np.max(np.abs(I_nA))
    hard_clip_nA = hard_clip_A * 1e9
    target_nA = I_target_A * 1e9
    if peak_nA == 0:
        raise RuntimeError("Waveform peak is zero.")
    if peak_nA > hard_clip_nA:
        raise RuntimeError(f"Waveform peak too large: {peak_nA:.3e} nA (>{hard_clip_nA:.3e} nA).")
    if not np.isclose(peak_nA, target_nA):
        scale = target_nA / peak_nA
        I_nA = I_nA * scale
        print(f"[prepare] scaled waveform by {scale:.3e} -> target peak {target_nA:.0f} nA")
    I_nA = np.clip(I_nA, -hard_clip_nA, hard_clip_nA)
    return t_ms, I_nA

COMMON_NODE_MECHS = ['axnode', 'AXNODE', 'axnodeNa', 'node']

def build_mrg_nodes_scaled(diameter_um=5.7,
                           num_nodes=21,
                           node_len_um=2.0,
                           internode_len_um=100.0,
                           gnabar_ref_diam=5.7,
                           gnabar_alpha=0.8,
                           gnabar_base_override=None):
    nodes=[]; internodes=[]
    for i in range(num_nodes):
        s = h.Section(name=f'node_{i}')
        s.L = node_len_um; s.diam = diameter_um; s.nseg = 1
        inserted=[]
        for mech in COMMON_NODE_MECHS:
            try:
                s.insert(mech); inserted.append(mech)
            except Exception:
                pass
        if 'hh' not in inserted:
            try:
                s.insert('hh'); inserted.append('hh')
            except Exception:
                pass
        nodes.append(s)
        if i < num_nodes-1:
            it = h.Section(name=f'int_{i}')
            it.L = internode_len_um; it.diam = diameter_um; it.nseg = 1
            try:
                it.insert('pas')
            except Exception:
                pass
            internodes.append(it)
    for i in range(len(internodes)):
        internodes[i].connect(nodes[i], 1.0)
        nodes[i+1].connect(internodes[i], 1.0)
    sample_seg = nodes[0](0.5)
    node_mech_name = None
    for nm in COMMON_NODE_MECHS:
        if hasattr(sample_seg, nm):
            node_mech_name = nm; break
    if node_mech_name is None and hasattr(sample_seg, 'hh'):
        node_mech_name = 'hh'
    if node_mech_name is not None:
        seg_mech = getattr(sample_seg, node_mech_name)
        possible_names = ['gnabar','gbar','gna','g_na','g_Na','gNa']
        gn_attr = None
        for name in possible_names:
            if hasattr(seg_mech, name):
                gn_attr = name; break
        if gn_attr is None:
            attrs = [a for a in dir(seg_mech) if not a.startswith('_')]
            numeric_attrs = []
            for a in attrs:
                try:
                    v = getattr(seg_mech, a)
                    if isinstance(v, (int, float)):
                        numeric_attrs.append((a, float(v)))
                except Exception:
                    pass
            cand = None
            for a,_ in numeric_attrs:
                if 'g' in a.lower() and ('na' in a.lower() or 'bar' in a.lower()):
                    cand = a; break
            if cand is None and numeric_attrs:
                cand = numeric_attrs[0][0]
            gn_attr = cand
        if gn_attr is not None:
            try:
                cur_val = float(getattr(seg_mech, gn_attr))
            except Exception:
                cur_val = None
            if cur_val is None:
                print("[build_mrg_nodes_scaled] could not read current gnabar value; skipping scaling.")
            else:
                if gnabar_base_override is not None:
                    base = float(gnabar_base_override)
                else:
                    base = cur_val / ((diameter_um / gnabar_ref_diam) ** gnabar_alpha) if gnabar_ref_diam != 0 else cur_val
                new_density = base * ((diameter_um / gnabar_ref_diam) ** gnabar_alpha)
                applied = 0
                for s in nodes:
                    seg = s(0.5)
                    try:
                        mech = getattr(seg, node_mech_name)
                        setattr(mech, gn_attr, new_density)
                        applied += 1
                    except Exception:
                        pass
                print(f"[build_mrg_nodes_scaled] set '{gn_attr}' on node mech '{node_mech_name}' to {new_density:.3e} (alpha={gnabar_alpha}) on {applied} nodes.")
        else:
            print("[build_mrg_nodes_scaled] No gnabar-like attribute detected; skipping scaling.")
    else:
        print("[build_mrg_nodes_scaled] No node mechanism object detected; skipping scaling.")
    diag = {'num_nodes': num_nodes, 'node_L': node_len_um, 'internode_L': internode_len_um}
    return nodes, internodes, diag

def _compute_internode_len(diameter_um):
    if FIXED_INTERNODE_FOR_DEBUG is not None:
        return float(FIXED_INTERNODE_FOR_DEBUG)
    internode = max(50.0, 100.0 * float(diameter_um))
    internode = min(internode, INTERNODE_CAP_UM)
    return internode

def run_and_detect_AP(nodes, t_ms, I_nA_wave, detect_idx=detect_node_idx):
    stim = h.IClamp(nodes[0](0.5))
    stim.delay = 0.0; stim.dur = 1e9
    vec_t = h.Vector(t_ms.tolist()); vec_i = h.Vector(I_nA_wave.tolist())
    vec_i.play(stim._ref_amp, vec_t)
    rec_t = h.Vector()
    rec_v_stim = h.Vector(); rec_v_det = h.Vector()
    rec_t.record(h._ref_t)
    rec_v_stim.record(nodes[0](0.5)._ref_v)
    rec_v_det.record(nodes[min(detect_idx, len(nodes)-1)](0.5)._ref_v)
    h.finitialize(-65.0); h.continuerun(t_ms[-1] + 5.0)
    v_stim = np.array(rec_v_stim); v_det = np.array(rec_v_det); time = np.array(rec_t)
    vm_max_stim = v_stim.max() if len(v_stim) > 0 else -999
    vm_max_det  = v_det.max()  if len(v_det)  > 0 else -999
    dvdt_stim = np.max(np.diff(v_stim)/np.diff(time)) if len(time) > 1 else 0.0
    dvdt_det  = np.max(np.diff(v_det)/np.diff(time))  if len(time) > 1 else 0.0
    ap_stim = (vm_max_stim > 0.0) or (dvdt_stim > 15.0)
    ap_det  = (vm_max_det  > 0.0) or (dvdt_det  > 15.0)
    return (ap_stim, ap_det), time, (v_stim, v_det)

def threshold_search(nodes_factory, t_ms, I_nA_unit, low=0.0,
                     max_factor=THRESH_MAX_FACTOR, tol=THRESH_TOL,
                     max_binary_iters=THRESH_BINARY_ITERS, debug=False):
    A = 1e-6 if low == 0 else float(low)
    found = False
    ap_stim_final = False
    ap_det_final = False
    while A <= max_factor:
        nodes, ints, diag = nodes_factory()
        (ap_stim, ap_det), _, _ = run_and_detect_AP(nodes, t_ms, I_nA_unit * A)
        if debug:
            print(f"[exp-search] A={A:.3g} -> ap_stim={ap_stim}, ap_det={ap_det}, internode={diag.get('internode_L')}")
        if ap_stim:
            found = True
            hi = A
            lo = max(A/2.0, 1e-12)
            break
        A *= 2.0
    if not found:
        if debug: print("[threshold_search] no AP found up to max_factor")
        return None, None, False, False
    for _ in range(max_binary_iters):
        mid = 0.5 * (lo + hi)
        nodes, ints, diag = nodes_factory()
        (ap_stim, ap_det), _, _ = run_and_detect_AP(nodes, t_ms, I_nA_unit * mid)
        if ap_stim:
            hi = mid
        else:
            lo = mid
        if hi > 0 and (hi - lo) / hi < tol:
            break
    A_th = hi
    try:
        Q_th = np.trapz((I_nA_unit * 1e-9) * A_th, t)
    except Exception:
        Q_th = np.trapz((I_nA_unit * 1e-9) * A_th, np.array(t_ms) * 1e-3)
    nodes, ints, diag = nodes_factory()
    (ap_stim_final, ap_det_final), _, _ = run_and_detect_AP(nodes, t_ms, I_nA_unit * A_th)
    if debug:
        print(f"[threshold_search] A_th={A_th:.6g}, Q_th={Q_th*1e9:.6g} nC, ap_stim={ap_stim_final}, ap_det={ap_det_final}, internode={diag.get('internode_L')}")
    return A_th, Q_th, ap_stim_final, ap_det_final

def main():
    try:
        load_mech_lib_safe(LIB_PATH)
    except Exception as e:
        print("Load failed:", e); return

    global t
    t = np.arange(0, t_total + dt_wave, dt_wave)

    results_all = {}

    for phase_order in PHASE_ORDERS:
        # set CV polarity to match CC phase order:
        # if CC is 'anodic' first, use positive V0 (anodic leading at electrode)
        # if CC is 'cathodic' first, use negative V0
        V0_sign = V0 if phase_order.lower().startswith('a') else -V0
        Vfunc = rectangular_pulse(V0_sign, t_on, t_off)
        I_raw_cv = generate_eti_current(Vfunc, t, R_s=100.0, R_ct=1e6, C_dl=1e-6)
        t_ms, I_nA_cv = prepare_waveform_for_neuron(t, I_raw_cv, I_target_A=I_target_A)

        I_raw_cc, ph_times = biphasic_current_waveform(I_target_A, BIP_DUR_CATH,
                                                       dur_anod_s=BIP_DUR_ANOD,
                                                       ipg_s=BIP_IPG,
                                                       tvec=t,
                                                       start_time=t_on,
                                                       phase_first=phase_order)
        mask_p1 = (t >= ph_times[0]) & (t < ph_times[1])
        mask_p2 = (t >= ph_times[2]) & (t < ph_times[3])
        Q_p1 = np.trapz(I_raw_cc[mask_p1], t[mask_p1]) if mask_p1.any() else 0.0
        Q_p2 = np.trapz(I_raw_cc[mask_p2], t[mask_p2]) if mask_p2.any() else 0.0
        print(f"[CC biphasic {phase_order}] Phase1 Q={Q_p1:.3e} C, Phase2 Q={Q_p2:.3e} C, net={Q_p1+Q_p2:.3e} C")

        t_ms_cc, I_nA_cc = prepare_waveform_for_neuron(t, I_raw_cc, I_target_A=I_target_A)

        I_peak_cv_A = float(np.max(np.abs(I_nA_cv))) * 1e-9
        I_peak_cc_A = float(np.max(np.abs(I_nA_cc))) * 1e-9

        plt.figure(figsize=(6,3))
        plt.plot(t_ms, I_nA_cv*1e-3, label='CV (ETI) µA')
        plt.plot(t_ms_cc, I_nA_cc*1e-3, ls='--', label=f'CC biphasic ({phase_order}) µA')
        plt.xlabel('Time (ms)'); plt.ylabel('Current (µA)'); plt.legend(); plt.title(f'Waveforms: CV vs CC ({phase_order})'); plt.grid(True); plt.show()

        results = {}
        for alpha in ALPHAS:
            print(f"\n=== sweep alpha = {alpha} phase_order = {phase_order} ===")
            results[alpha] = {}
            for d in DIAMETERS:
                print(f" Testing diameter {d} µm ...", end=' ')
                def nodes_factory_local(dlocal=d, alpha_local=alpha):
                    internode_len = _compute_internode_len(dlocal)
                    nodes, ints, diag = build_mrg_nodes_scaled(dlocal, num_nodes=21, node_len_um=2.0,
                                                               internode_len_um=internode_len,
                                                               gnabar_ref_diam=5.7,
                                                               gnabar_alpha=alpha_local,
                                                               gnabar_base_override=None)
                    diag['diameter_um'] = dlocal
                    return nodes, ints, diag

                try:
                    A_th_cv, Q_th_cv, ap_stim_cv, ap_det_cv = threshold_search(nodes_factory_local, t_ms, I_nA_cv, low=0.0, debug=False)
                except Exception as e:
                    print("-> ERROR CV:", e); A_th_cv, Q_th_cv, ap_stim_cv, ap_det_cv = None, None, False, False

                try:
                    A_th_cc, Q_th_cc, ap_stim_cc, ap_det_cc = threshold_search(nodes_factory_local, t_ms_cc, I_nA_cc, low=0.0, debug=False)
                except Exception as e:
                    print("-> ERROR CC:", e); A_th_cc, Q_th_cc, ap_stim_cc, ap_det_cc = None, None, False, False

                Ith_cv_pA = (A_th_cv * I_peak_cv_A * 1e12) if A_th_cv is not None else np.nan
                Ith_cc_pA = (A_th_cc * I_peak_cc_A * 1e12) if A_th_cc is not None else np.nan

                diag_internode = None
                try:
                    _, _, diag_tmp = nodes_factory_local()
                    diag_internode = diag_tmp.get('internode_L')
                except Exception:
                    diag_tmp = {'internode_L': None}

                print(f"-> diam={d} internode={diag_internode} | CV: A_th={A_th_cv} ap_stim={ap_stim_cv} ap_det={ap_det_cv} Ith={Ith_cv_pA:.3f} pA | CC: A_th={A_th_cc} ap_stim={ap_stim_cc} ap_det={ap_det_cc} Ith={Ith_cc_pA:.3f} pA")

                results[alpha][d] = {
                    'A_th_cv': A_th_cv, 'Q_th_cv': Q_th_cv, 'Ith_cv_pA': Ith_cv_pA, 'ap_stim_cv': ap_stim_cv, 'ap_det_cv': ap_det_cv,
                    'A_th_cc': A_th_cc, 'Q_th_cc': Q_th_cc, 'Ith_cc_pA': Ith_cc_pA, 'ap_stim_cc': ap_stim_cc, 'ap_det_cc': ap_det_cc
                }

        results_all[phase_order] = results

        plt.figure(figsize=(8,4))
        for alpha in ALPHAS:
            Avals_cv = [results[alpha][d]['A_th_cv'] if results[alpha][d]['A_th_cv'] is not None else np.nan for d in DIAMETERS]
            Avals_cc = [results[alpha][d]['A_th_cc'] if results[alpha][d]['A_th_cc'] is not None else np.nan for d in DIAMETERS]
            plt.plot(DIAMETERS, Avals_cc, marker='o', label=f'CC alpha={alpha}')
            plt.plot(DIAMETERS, Avals_cv, marker='x', ls='--', label=f'CV alpha={alpha}')
        plt.xlabel('diameter (µm)'); plt.ylabel('A_th (scale)'); plt.title(f'A_th: CC vs CV ({phase_order})'); plt.legend(); plt.grid(True); plt.show()

        plt.figure(figsize=(8,4))
        for alpha in ALPHAS:
            Qvals_cc = [ (results[alpha][d]['Q_th_cc']*1e9) if results[alpha][d]['Q_th_cc'] is not None else np.nan for d in DIAMETERS]
            Qvals_cv = [ (results[alpha][d]['Q_th_cv']*1e9) if results[alpha][d]['Q_th_cv'] is not None else np.nan for d in DIAMETERS]
            plt.plot(DIAMETERS, Qvals_cc, marker='o', label=f'CC alpha={alpha}')
            plt.plot(DIAMETERS, Qvals_cv, marker='x', ls='--', label=f'CV alpha={alpha}')
        plt.xlabel('diameter (µm)'); plt.ylabel('Q_th (nC)'); plt.title(f'Q_th: CC vs CV ({phase_order})'); plt.legend(); plt.grid(True); plt.show()

        plt.figure(figsize=(8,4))
        for alpha in ALPHAS:
            Ith_cc = [ results[alpha][d]['Ith_cc_pA'] if not np.isnan(results[alpha][d]['Ith_cc_pA']) else np.nan for d in DIAMETERS]
            Ith_cv = [ results[alpha][d]['Ith_cv_pA'] if not np.isnan(results[alpha][d]['Ith_cv_pA']) else np.nan for d in DIAMETERS]
            plt.plot(DIAMETERS, Ith_cc, marker='o', label=f'CC Ith (pA) alpha={alpha}')
            plt.plot(DIAMETERS, Ith_cv, marker='x', ls='--', label=f'CV Ith (pA) alpha={alpha}')
        plt.xlabel('diameter (µm)'); plt.ylabel('Ith (pA)'); plt.title(f'Ith: CC vs CV ({phase_order})'); plt.legend(); plt.grid(True); plt.show()

    print("\nDone. Summary keys in 'results_all' by phase order (anodic/cathodic).")

if __name__ == "__main__":
    main()
```
